```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
rng = np.random
learning_rate = 0.1
training_epochs = 10000
display_step = 500
batch_size = 100
```

```python
Z = pd.read_csv('./train.csv',header = None , usecols = [1,2,4,5,6,7])
Z = Z.drop(0, 0)
for i, array in enumerate(Z[5]):
    try:
        int(array)
    except:
        Z[5][i+1] = 29.7
X_train = np.array(Z.drop(1,1))
Y = np.array(Z[1])
for i,val in enumerate(X_train):
    Y[i] = float(Y[i])
    if(val[1] == 'male'):
        val[1] = 1
    else:
        val[1] = 0
    for k in range(len(val)):
        val[k] = float(val[k])
Y_train = []
for i in range(len(Y)):
    if(Y[i]==0):
        Y_train.append([1,0])
    else:
        Y_train.append([0,1])
        
DF = pd.read_csv('./test.csv',header = None , usecols = [1,3,4,5,6])
DF = DF.drop(0, 0)
for i, array in enumerate(DF[4]):
    try:
        int(array)
    except:
        DF[4][i+1] = 29.7
X_test = np.array(DF)
for i,val in enumerate(X_test):
    if(val[1] == 'male'):
        val[1] = 1
    else:
        val[1] = 0
    for k in range(len(val)):
        val[k] = float(val[k])
        
X_test = X_test.astype(float)

X_test = np.reshape(X_test,(418,5))
```

```python
X = tf.placeholder(tf.float32,[None,5])
Y = tf.placeholder(tf.float32,[None,2])

W = tf.Variable(tf.zeros([5,2]))
b = tf.Variable(tf.zeros([2]))

pred = tf.nn.softmax(tf.matmul(X,W)+b)

cross_entropy = tf.reduce_mean(-tf.reduce_sum(Y * tf.log(pred),reduction_indices = [1]))

optimizer = tf.train.AdamOptimizer(learning_rate).minimize(cross_entropy)

model = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(model)
    for epoch in range(training_epochs):
        total_batch = int(891/batch_size)
        X_batches = np.array_split(X_train, total_batch)
        Y_batches = np.array_split(Y_train, total_batch)
        for t in range(total_batch):
            batch_x, batch_y = X_batches[t], Y_batches[t]    
            c = sess.run([optimizer,cross_entropy],feed_dict = {X:batch_x , Y:batch_y})
            cost = c[1]/total_batch
        if (epoch+1) % display_step ==0:
            print("Epoch:{} cost :{}".format(epoch,cost))
    print("Optimization Finished")
    training_cost = sess.run(cross_entropy,feed_dict = {X:X_train , Y:Y_train})
    weight = sess.run(W)
    bias = sess.run(b)
    print("Final cost is:{},Weight = {} bias = {}".format(training_cost,weight,bias))
    correct_prediction = tf.equal(tf.argmax(pred, 1), tf.argmax(Y_train, 1))
    # Calculate accuracy
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
    print("Accuracy:", accuracy.eval({X: X_train, Y: Y_train})*100)
    global result 
    result = tf.argmax(pred, 1).eval({X: X_train, Y: Y_train})
    final = sess.run(pred,feed_dict = {X: X_test})
    final_pred = []
    for i in range(418):
        if(final[i][0]>final[i][1]):
            t = 0
            final_pred.append(t)
        elif(final[i][1]>final[i][0]):
            t = 1
            final_pred.append(t)
    myData = [["PassengerId", "Survived"]]
    for i in range(len(DF)):
        arr = [i+892, int(final_pred[i])]
        myData.append(arr)
    myFile = open('example3.csv', 'w+')
    with myFile:
        writer = csv.writer(myFile)
        writer.writerows(myData)
    print("WRITING COMPLETE!!!")
        
```
